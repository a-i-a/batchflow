```python
import sys
sys.path.append('../..')

from batchflow import Pipeline, B, C, V
from batchflow.opensets import MNIST
from batchflow.models.tf import ResNet18, ResNet34
from batchflow.research import Research, Option
```

```python
BATCH_SIZE=64
```

```python
model_config = {'inputs/images/shape': B('image_shape'),
                'inputs/labels/classes': 10,
                'initial_block/inputs': 'images'}
```

```python
mnist = MNIST()
train_root = mnist.train.p.run(BATCH_SIZE, shuffle=True, n_epochs=None, lazy=True)
```

```python
train_template = (Pipeline()
            .init_variable('loss', init_on_each_run=list)
            .init_model('dynamic', C('model'), 'conv', config=model_config)
            .to_array()
            .train_model('conv', fetches='loss',
                         feed_dict={'images': B('images'),
                                    'labels': B('labels')},
                         save_to=V('loss', mode='w'))
)
train_ppl = train_root + train_template
```

```python
grid = Option('model', [ResNet18, ResNet34])
```

```python
research = (Research()
            .add_pipeline(train_ppl, variables='loss', name='train')
            .add_grid(grid))

research.run(n_reps=2, n_iters=10, name='my_research', bar=True)
```

```python
results = research.load_results()
results
```

```python
research = (Research()
            .add_pipeline(train_ppl, variables='loss', execute='%2', name='train')
            .add_grid(grid))
research.run(n_reps=2, n_iters=10, name='my_research', bar=True)
```
